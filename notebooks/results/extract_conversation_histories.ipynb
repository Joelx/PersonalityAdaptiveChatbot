```python
input_file = 'logs_2.log'
output_file = 'parsed_histories_9.txt'
separator = '\n\n---BLOCK_SEPARATOR---\n\n'

def extract_blocks(input_file, output_file, separator):
    with open(input_file, 'r', encoding='utf-8') as infile:
        content = infile.read()
    
    blocks = []
    start_marker = 'Current Big 5 Personality traits: '
    end_marker = '> Finished chain.'
    
    start_index = content.find(start_marker)
    
    while start_index != -1:
        end_index = content.find(end_marker, start_index) + len(end_marker)
        blocks.append(content[start_index:end_index])
        start_index = content.find(start_marker, end_index)
    
    with open(output_file, 'w', encoding='utf-8') as outfile:
        outfile.write(separator.join(blocks))

extract_blocks(input_file, output_file, separator)
```

```python
def load_blocks(file_path, separator):
    with open(file_path, 'r') as file:
        content = file.read()
    return content.split(separator)

blocks = load_blocks(output_file, separator)
```
